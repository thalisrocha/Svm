## Preliminary questions

**1)** Show that the primal problem solved by the SVM can be rewritten as follows :

$$ \text{argmin}_{\mathbf{w} \in \mathcal{H}, w_0 \in \mathbb{R} } \left( \frac{1}{2}||\mathbf{w}||^2 + C \sum_{i=1}^n [ 1 - y_i ( \langle \mathbf{w}, \Phi(\mathbf{x_i}) \rangle + w_0 )]_+ \right) $$

<div class='alert alert-block alert-warning'>
            Answer:</div>


For our SVM, we have the primal problem under constraints:
$$
\left\{
\begin{array}{l}
    \displaystyle \mathop{\mathrm{argmin}}_{\mathbf{w} \in \mathcal{H}, \, w_0 \in \mathbb{R}, \, \xi \in \mathbb{R}^n} \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^n \xi_i \\
    \forall i, \quad \xi_i \ge 0 \\
    \forall i, \quad \xi_i \ge 1 - y_i(\langle \mathbf{w}, \Phi(\mathbf{x_i}) \rangle + w_0)
\end{array}
\right.
$$
Now we can substitute the optimal $\xi^*$ and get the function:
$$
\displaystyle \mathop{\mathrm{argmin}}_{\mathbf{w} \in \mathcal{H}, w_0 \in \mathbb{R}} \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^n \max(0, 1 - y_i(\langle \mathbf{w}, \Phi(\mathbf{x_i}) \rangle + w_0))
$$

The function $x \rightarrow [1 - x]_+ = \text{max}(0, 1-x)$ is called *Hinge* and we can use it to rewrite our problem and achieve the result:

$$ \text{argmin}_{\mathbf{w} \in \mathcal{H}, w_0 \in \mathbb{R} } \left( \frac{1}{2}||\mathbf{w}||^2 + C \sum_{i=1}^n [ 1 - y_i ( \langle \mathbf{w}, \Phi(\mathbf{x_i}) \rangle + w_0 )]_+ \right) $$

**2)** Explain the sentence : "an SVM minimizes the classification error using a convex upper bound". The function $x \rightarrow [1 - x]_+ = \text{max}(0, 1-x)$ is called *Hinge* (*charnière* en français). Explain the difference between the pivotal loss and the loss of binary classification.

<div class='alert alert-block alert-warning'>
            Answer:</div>

The sentence "an SVM minimizes the classification error using a convex upper bound" can be explained by the definition of SVM, the SVM optimization tries to find a hyperplane in a high-dimensional space that maximally separates the different classes of data, formulated as the convex optimization problem we saw in the last exercise, that uses a convex upper bound to minimize the classification error. In other words, the algorithm tries to find a hyperplane that correctly separates the data into different classes, while minimizing the classification error.

The hinge loss is a loss used in binary classification problems to quantify the error between the predicted and the true class which penalizes misclassifications more heavily than correct classifications.

The pivotal loss is another loss function, but it is not used to measure the error between the predicted and the true class, but to estimate the confidence intervals of the estimated parameters.




## Example of using the SVC class from scikit-learn

````python
#  Data Generation

def rand_gauss(n=100, mu=[1, 1], sigmas=[0.1, 0.1]):
    """ Sample n points from a Gaussian variable with center mu,
    and std deviation sigma
    """
    d = len(mu)
    res = np.random.randn(n, d)
    return np.array(res * sigmas + mu)


def rand_bi_gauss(n1=100, n2=100, mu1=[1, 1], mu2=[-1, -1], sigmas1=[0.1, 0.1],
                  sigmas2=[0.1, 0.1]):
    """ Sample n1 and n2 points from two Gaussian variables centered in mu1,
    mu2, with respective std deviations sigma1 and sigma2
    """
    ex1 = rand_gauss(n1, mu1, sigmas1)
    ex2 = rand_gauss(n2, mu2, sigmas2)
    y = np.hstack([np.ones(n1), -1 * np.ones(n2)])
    X = np.vstack([ex1, ex2])
    ind = np.random.permutation(n1 + n2)
    return X[ind, :], y[ind]
````

````python
# This is an example of using SVC for data generated with the above function
n1 = 20
n2 = 20
mu1 = [1., 1.]
mu2 = [-1., -1.]
sigma1 = [0.9, 0.9]
sigma2 = [0.9, 0.9]
X1, y1 = rand_bi_gauss(n1, n2, mu1, mu2, sigma1, sigma2)

X_train = X1[::2]
Y_train = y1[::2].astype(int)
X_test = X1[1::2]
Y_test = y1[1::2].astype(int)

# fit the model with linear kernel
clf = SVC(kernel='linear')
clf.fit(X_train, Y_train)

# predict labels for the test data base
y_pred = clf.predict(X_test)

# check your score
score_train = clf.score(X_train, Y_train)
score_test = clf.score(X_test, Y_test)
print('Training score : %s' % score_train)
print('Testing score : %s' % score_test)

# display the points
plt.figure(1, figsize=(5, 5))
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    clf,
    X1,
    plot_method="contour",
    colors="k",
    levels=[0],
    alpha=0.5,
    ax=ax,
)
ax.scatter(X_train[:, 0], X_train[:, 1], c=Y_train)
ax.scatter(X_test[:, 0], X_test[:, 1], c=Y_test, cmap=plt.cm.Paired)
plt.title('First data set')
plt.axis('equal')
plt.show()
````

**4)** Since the probability distributions are known, numerically estimate the Bayes risk.

<div class='alert alert-block alert-info'>
            Code:</div>

````python
from scipy.stats import multivariate_normal

mu1 = np.mean(X[pos,:], axis=0)
mu2 = np.mean(X[neg,:], axis=0)
sigma1 = np.cov(X[pos,:].T)
sigma2 = np.cov(X[neg,:].T)

# Do a Monte-Carlo estimation of the Bayes Risk (given the gaussian distributions you used to generate data)
# You need to use the imported class from scipy to generate a large number of samples which you will use
# to approximate the integral of the Bayes risk
n_mc = 1000 # Repeat this n_mc times - enough to approximate
expectation = 0
for i in range(n_mc):
    rand = np.random.randint(2) # Random binary choice: will the point be generated from the first or second gaussian ?
    if rand==0:
        # First case: y = 0
        x = np.random.multivariate_normal(mean=mu1, cov=sigma1)
    else:
        # Second case: y = 1
        x = np.random.multivariate_normal(mean=mu2, cov=sigma2)

    # You have to compute the conditional posterior probability of x given the 2 gaussians
    # Use the multivariate_normal.pdf() method !
    p1 = multivariate_normal.pdf(x = x, mean = mu1, cov = sigma1)
    p2 = multivariate_normal.pdf(x = x, mean = mu2, cov = sigma2)
    # Compute the risk from these and add it to the total
    risk = min(p1, p2)
    risk = 1 - risk if rand == 0 else risk

    expectation += min(p1/(p1+p2), p2/(p1+p2))

expectation /= n_mc

print(f'Estimated Bayes risk: {np.around(expectation, 3)}')
print(f'Estimated Bayes accuracy: {1 - np.around(expectation, 3)}')
````

**Answer:**

For the **rbf** kernel with **gamma=1** and **C=1000.0**, the training score is 0.976 and the testing score is 0.84. This suggests that the model is performing very well on the training data but is not able to generalize to new data, indicating overfitting. For the same kernel with gamma=10 and C=31.622776601683793, the training score is 1.0 and the testing score is 0.852. Again, this suggests overfitting. Also for the gaussian kernel with gamma=100 and C=0.001, the training score is 1.0 and the testing score is 0.82.

Gamma is a hyperparameter that determines the width of the Gaussian kernel in the radial basis function (RBF) kernel. A large value of gamma results in a narrower kernel, which makes the decision boundary more complex and able to capture finer details in the data. In general, a smaller value of gamma may lead to underfitting, while a larger value may lead to overfitting.

C is a hyperparameter that controls the trade-off between achieving a low training error and a low testing error. A larger value of C forces the model to correctly classify more training examples, leading to a smaller margin and a more complex model, which may result in overfitting.

In the given data, it is possible to see that the cases that we have overfitting C or gamma have big values (not necessarily at the same time), leading the model to overfitting.  It is important to carefully select appropriate values of gamma and C to obtain good generalization performance.

````python
####################################################################
# Qualitative evaluation of the predictions using matplotlib

def plot_gallery(images, titles, n_row=3, n_col=4):
    """Helper function to plot a gallery of portraits"""
    pl.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    pl.subplots_adjust(bottom=0, left=.01, right=.99, top=.90,
                       hspace=.35)
    for i in range(n_row * n_col):
        pl.subplot(n_row, n_col, i + 1)
        pl.imshow(images[i])
        pl.title(titles[i], size=12)
        pl.xticks(())
        pl.yticks(())


def title(y_pred, y_test, names):
    pred_name = names[int(y_pred)].rsplit(' ', 1)[-1]
    true_name = names[int(y_test)].rsplit(' ', 1)[-1]
    return 'predicted: %s\ntrue:      %s' % (pred_name, true_name)

# This will just show some examples with their associated prediction - nothing to change
prediction_titles = [title(y_pred[i], y_test[i], names)
                     for i in range(y_pred.shape[0])]

plot_gallery(images_test, prediction_titles)
pl.show()
````
